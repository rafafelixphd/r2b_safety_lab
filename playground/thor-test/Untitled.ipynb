```python
from pyvirtualdisplay import Display
from ai2thor.controller import Controller
from PIL import Image

# 1. Start the virtual display (if not already running)
try:
    v_display = Display(visible=0, size=(1024, 768))
    v_display.start()
except Exception as e:
    print(f"Display already running or error: {e}")

# 2. Initialize Controller
controller = Controller(scene="FloorPlan1", width=640, height=480)

# 3. Take an action
event = controller.step(action="RotateRight")

# 4. Display with Pillow
# AI2-THOR returns a numpy array in RGB at event.frame
img = Image.fromarray(event.frame)
img.show() # This opens a window if on desktop, but in Jupyter, just do:
img
```
